```python
to_do_list = []
while True :
  print("\n1. View Task")
  print("2. Add Task")
  print("3.Mark Task as Done")
  print("4. Delete Task")
  print("5. Exit")
  choice = input("Enter your choice(1-5) :")
  if choice == '1':
    if not  to_do_list :
      print("Your to_do_list is empty")
    else :
      print("your to_do_list is :")
      for index , task in enumerate(to_do_list , start=1) :
        print(f"{index}.{task}")
  elif choice == '2' :
    task = input("Enter the task you want to add :")
    to_do_list.append(task)
    print(f"Task '{task}' added successfully")
  elif choice == '3' :
    if not to_do_list :
      print("Your to_do_list is empty")
    else :
      print("\n Tasks:")
      for index, task in enumerate(to_do_list, start=1) :
        print(f"{index}.{task}")
      done_index = int(input("Enter the task number you want to mark as done :"))
      if 1 <=done_index <=len(to_do_list) :
        done_task = to_do_list.pop(done_index - 1)
        print(f"Task '{done_task}' marked as done")
      else :
        print("Invalid task number")
  elif choice == '4' :
    if not to_do_list :
      print("Your to_do_list is empty")
    else :
      print("\n Tasks:")
      for index, task in enumerate(to_do_list, start=1) :
        print(f"{index}.{task}")
      delete_index = int(input("Enter the task number you want to delete :"))
      if 1 <=delete_index <=len(to_do_list) :
        deleted_task = to_do_list.pop(delete_index - 1)
        print(f"Task '{deleted_task}' deleted successfully")
  elif choice == '5' :
    print("Exiting the to-do list application. Goodbye!")
    break
  else :
    print("Invalid choice. Please enter a number between 1 and 5.")
```
